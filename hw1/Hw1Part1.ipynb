```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## 3 Simple Functions (f, g and h)

```
#Defining 3 functions (f, g and h)
def f(x, y):
    return (x ** 2) - (y ** 2)

def g(x, y):
    return (x ** 4 - 2*x ** 2) - (y ** 4 - 2*y ** 2)

def h(x, y):
    return -(x ** 2) - (y ** 2)
 
```

## Graphing f(x, y) and its Saddle Point in Green

```
x = np.linspace(-5, 5, 30) #Range and intervals
y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter([0], [0], [0], color="g", s=30) #plot saddle point
```

### Contour of f(x, y)

```
x = np.linspace(-5, 5, 50) #Range and Interval
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.plot(0, 0, 'go') #plot saddle point
```

## Graphing g(x, y) with Local Min (Blue), Local Max (Red), Saddle Point (Green)

```
x = np.linspace(-1, 1, 30) #Range and intervals
y = np.linspace(-1, 1, 30)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 25, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.scatter([0], [0], [0], color="g", s=50) #plot saddle point
ax.scatter([0], [1], [1], color="r", s=10) #plot Local Max
ax.scatter([0], [-1], [1], color="r", s=10) #plot Local Max
ax.scatter([1], [0], [-1], color="b", s=10) #plot Local Min
ax.scatter([-1], [0], [-1], color="b", s=10) #plot Local Min
```

### Contour of g(x, y)

```
x = np.linspace(-1.5, 1.5, 50)
y = np.linspace(-1.5, 1.5, 40)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.plot( 0,  0, 'go') #plot saddle point
plt.plot( 0,  1, 'ro') #plot Local Max
plt.plot( 0, -1, 'ro') #plot Local Max
plt.plot( 1,  0, 'bo') #plot Local Min
plt.plot(-1,  0, 'bo') #plot Local Min
```

## Graphing h(x, y) and its Absolute Max (Red)

```
x = np.linspace(-5, 5, 30) #Range and intervals
y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x, y)
Z = h(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter([0], [0], [0], color="r", s=20) #plot absolute max
```

```
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)

X, Y = np.meshgrid(x, y)
Z = h(X, Y)
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.plot(0, 0, 'ro') #plot absolute max
```
